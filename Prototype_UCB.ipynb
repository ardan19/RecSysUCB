# UPPER CONFIDENCE BOUND

```python
music_data = dataframe_lagu
user_behavior_data = datapengguna
class UCBMusicRecommender:
    def __init__(self, num_songs):
        self.num_songs = num_songs
        self.num_selections = np.zeros(num_songs)
        self.sum_rewards = np.zeros(num_songs)

    def recommend_song(self, iteration):
        if iteration < self.num_songs:
            return iteration
        else:
            ucb_values = np.zeros(self.num_songs)
            for i in range(self.num_songs):
                if self.num_selections[i] > 0:
                    average_reward = self.sum_rewards[i] / self.num_selections[i]
                    delta_i = np.sqrt(2 * np.log(iteration) / self.num_selections[i])
                    ucb_values[i] = average_reward + delta_i
                else:
                    ucb_values[i] = float('inf')
            return np.argmax(ucb_values)

    def update_parameters(self, song_index, reward):
        self.num_selections[song_index] += 1
        self.sum_rewards[song_index] += reward

num_songs = len(music_data)
ucb_recommender = UCBMusicRecommender(num_songs)

num_iterations = len(datapengguna['nama_pengguna'])
print("Lagu yang direkomendasikan")
for i in range(num_iterations):
    recommended_song_index = ucb_recommender.recommend_song(i+1)
    recommended_song = music_data.iloc[recommended_song_index]
    user_action = user_behavior_data.sample(n=1).iloc[0]
    song_index = music_data[music_data['track_id'] == user_action['track_id']].index[0]
    reward = 1 if user_action['action'] == 'like' else 0
    ucb_recommender.update_parameters(song_index, reward)

    print("===============================")
    print("Judul Lagu:", recommended_song['track_name'])
    print("Artis:", recommended_song['artists'])
    print("Genre:", recommended_song['track_genre'])
```

```python
import numpy as np
import pandas as pd
from matplotlib_venn import venn2
import matplotlib.pyplot as plt

# Dikosongkan untuk menyederhanakan contoh
# Anggap music_data dan user_behavior_data didefinisikan di tempat lain

music_data = dataframe_lagu
user_behavior_data = datapengguna

class UCBMusicRecommender:
    def __init__(self, num_songs):
        self.num_songs = num_songs
        self.num_selections = np.zeros(num_songs)
        self.sum_rewards = np.zeros(num_songs)

    def recommend_song(self, iteration):
        if iteration < self.num_songs:
            return iteration
        else:
            ucb_values = np.zeros(self.num_songs)
            for i in range(self.num_songs):
                if self.num_selections[i] > 0:
                    average_reward = self.sum_rewards[i] / self.num_selections[i]
                    delta_i = np.sqrt(2 * np.log(iteration) / self.num_selections[i])
                    ucb_values[i] = average_reward + delta_i
                else:
                    ucb_values[i] = float('inf')
            return np.argmax(ucb_values)

    def update_parameters(self, song_index, reward):
        self.num_selections[song_index] += 1
        self.sum_rewards[song_index] += reward

num_songs = len(music_data)
ucb_recommender = UCBMusicRecommender(num_songs)

num_iterations = len(datapengguna['nama_pengguna'])

# Arrays to store recommendation outcomes
recommended_songs = []
liked_songs = []

print("Lagu yang direkomendasikan")
for i in range(num_iterations):
    recommended_song_index = ucb_recommender.recommend_song(i+1)
    recommended_song = music_data.iloc[recommended_song_index]
    recommended_songs.append(recommended_song['track_name'])

    user_action = user_behavior_data.sample(n=1).iloc[0]
    song_index = music_data[music_data['track_id'] == user_action['track_id']].index[0]
    reward = 1 if user_action['action'] == 'like' else 0
    ucb_recommender.update_parameters(song_index, reward)

    if reward == 1:
        liked_songs.append(recommended_song['track_name'])

    print("===============================")
    print("Judul Lagu:", recommended_song['track_name'])
    print("Artis:", recommended_song['artists'])
    print("Genre:", recommended_song['track_genre'])
```

```python
# Create sets
all_recommended_songs = set(recommended_songs)
liked_songs_set = set(liked_songs)

# Create Venn diagram
venn2([all_recommended_songs, liked_songs_set], ('Musik Direkomendasi', 'Musik Disukai'))
plt.title('Diagram Venn antara musik yang direkomendasi dan disukai')
plt.show()
```
